## Dataset Analysis and Visualization

In this notebook, we analyze the CIFAR-10 dataset, display the first 5 rows, and generate 100 real-world questions solvable using Numpy, Pandas, Matplotlib, and Seaborn.

### Step 1: Load the Dataset and Display the First 5 Rows


```python
import tensorflow as tf
import matplotlib.pyplot as plt

# Load CIFAR-10 dataset
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.cifar10.load_data()

# Define class names
class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer',
               'dog', 'frog', 'horse', 'ship', 'truck']

# Display the first 5 images with labels
plt.figure(figsize=(10, 2))
for i in range(5):
    plt.subplot(1, 5, i + 1)
    plt.imshow(x_train[i])
    plt.title(class_names[y_train[i][0]])
    plt.axis('off')
plt.show()

# Dataset Shape
print(f"Training Data Shape: {x_train.shape}, Labels Shape: {y_train.shape}")
print(f"Test Data Shape: {x_test.shape}, Labels Shape: {y_test.shape}")
```
